Manipulation de tableaux

```python
A = np.zeros((3,2))
B = np.ones((3, 2))
```

```python
print(A)
```

```python
print(B)
```

```python
# 1) assembler des array avec hstack et vstack pour horizontale et verticale
C = np.hstack((A, B))
C
# pour coller de façon horizontale
```

```python
C.shape
```

```python
D = np.vstack((A, B))
D
```

```python
D.shape
```

```python
# avec concatenate on colle et en 3eme argument on dit sur quel axe 
E = np.concatenate((A, B), axis=0)
E
```

```python
F = np.concatenate((A, B), axis=1)
F
```

```python
# maintenant -> fct reshape -> remanipuler la forme d'un array pour lui en donner une nouvelle
# attention -> ne fonctionne que si le nombre d'élément forme initial = nbr élément forme final
```

```python
D.shape
```

```python
D.size
```

```python
# on peut redimensionner mais ça doit toujours faire 12
# 6 x 2 ou 3 x 4 ou 4 x 3 ou 12 x 1 etc
```

```python
E = D.reshape((3, 4))
E
```

```python
# Exemple de l'utilité de reshape
G = np.array((1, 2, 3))
G.shape
```

```python
# le problème c'est qu'il n'est pas écrit (3, 1) -> ça peut poser problème dans certains algo
# si on veut écrire un 1 à la place de rien -> on utilise reshape
F = np.array((1,2,3))
F = F.reshape((F.shape[0], 1))
F.shape
 # on écrit F.shape[0] car c'est un tuple -> accès par index -> et c'est plus dynamique
```

```python
# De même si on veut l'enlever -> methode squeeze
F = F.squeeze()
F.shape
```
