Exercice -> zoomer de 1/4 vers le centre de la photo -> bonus : gérer la luminosité en réduisant pixels trop sombres et augmentant pixels trop clairs

```python
# Créer une variable hauteur et une variable largeur
h = face.shape[0]
w = face.shape[1]
print(h)
```

```python
print(w)
```

```python
# on va créer un nouveau tableau avec un subset
# quand on utilise la double division -> numpy retourne un entier -> c'est bien car ici besoin d'index
zoom_face = face[h//4 : -h//4, w//4 : -w//4]
plt.imshow(zoom_face, cmap=plt.cm.gray)
plt.show()
```

```python
# en plus on pourrait accentuer les pixels proche du blanc en les saturant à une valeur de 255
zoom_face = face[h//4 : -h//4, w//4 : -w//4]
zoom_face[zoom_face > 150] = 255
plt.imshow(zoom_face, cmap=plt.cm.gray)
plt.show()
```
