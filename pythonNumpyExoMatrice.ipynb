```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
np.random.seed(0)
A = np.random.randint(0, 100, [10, 5])
A
```

```python
# On va faire du broadcasting
D = (A - A.mean(axis=0)) / A.std(axis=0)
D     
```

```python
# Toutes les colonnes ont été standardisées selon la formule au dessus
# ça veut dire que !!!!!!!! = 
# La moyenne de chaque colone = 0
# l'écart type de chaque colone = 1
```

```python
D.mean(axis=0)
```

```python
# -> très proche de 0
```

```python
D.std(axis=0)
```

```python
# -> bien = à 1
```
